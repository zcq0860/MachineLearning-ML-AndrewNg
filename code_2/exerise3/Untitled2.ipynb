```python
#引入使用的模块
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
#处理并查看数据
data=loadmat('F:\\Machinelearning\data\ex3data1')
print(data)
```

```python
X=data['X']
y=data['y']
print(X.shape,y.shape)
```

```python
#定义logistic函数
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
#给X加常数列，处理偏转项；定义theta
m,d=X.shape
X=np.insert(X,0,1,axis=1)
theta=np.zeros(d+1)
y=y.reshape((m,))
print(X.shape,y.shape,theta.shape)
print(X)
```

```python
def regularized_Cost(X,y,theta,regular_param):
    m=X.shape[0]
    cost=-np.mean(y*np.log(sigmoid(X@theta))+(1-y)*np.log(1-sigmoid(X@theta)))
    reg_term=(theta[1:].T@theta[1:])/(2*m)
    return cost+reg_term
print(regularized_Cost(X,y,theta,1))
```

```python
def reg_Gradient(X,y,theta,regular_param):
    m=X.shape[0]
    delta_Vector=sigmoid(X@theta)-y
    gradient=(X.T@delta_Vector)/m
    theta1=theta[1:]
    theta1=np.insert(theta1,0,0)
    reg_term=theta1*regular_param/m
    return gradient+reg_term,reg_term.shape,gradient.shape
print(reg_Gradient(X,y,theta,1))
```

```python
from scipy.optimize import minimize
def one_vs_all(X,y,classNum,Reg_Param):
    all_theta=np.zeros((classNum,d+1))
    for i in range(classNum):
        y1=[1 if i+1==label else 0 for label in y]
        fmin=minimize(fun=regularized_Cost,x0=theta,args=(X,y,Reg_Param),method='TNC',jac=reg_Gradient)
        all_theta[i,:]=fmin.x
    return all_theta
print(one_vs_all(X,y,10,1))
```
